```python

import scipy.stats as stats

#P(10 <= X <= 14) = P(X <= 14) - P(X >= 10)
mu = 12
sigma = 2
prob = stats.norm.cdf(14, mu, sigma) - stats.norm.cdf(10, mu, sigma)
print(prob)
```



Many times in data analysis, we need a method that relies on repeated random sampling to obtain numerical results. The underlying concept is to use randomness to solve problems. In fact, this is a mathematical technique, which is used to estimate the possible outcomes of an uncertain event and is called the `Monte Carlo` Method.

Consider that we roll a die 10 times and want to know the probability of getting more than 3 times of even numbers. This is a problem for the Binomial distribution, but suppose we don’t know anything about Binomial distribution. We can easily solve this problem with a simulation or random number generation, called Monte Carlo Simulation.




The Monte Carlo Method uses random numbers to simulate some process. Here, the process is rolling a die 20 times. Assume the die is fair. What is the probability of success or getting an even number in rolling the die once?

```python
import numpy as np
import random

total_outcomes = 6
even_outcomes = 3

num_simulation = 20
even_count = 0
results = []
for _ in range(num_simulation):
    roll = random.randint(1, total_outcomes)
    results.append(roll)

    if roll % 2 == 0:
        even_count = 1

print(results)
print(even_count)


Monte_Carlo_Probabilty = even_count / num_simulation
print(Monte_Carlo_Probabilty)
```



Define a `function` named `one_trial`, that simulates a single round of rolling a die 20 times and returns true if the number of even numbers is > 5.

```python
import random

def one_trial():
    even_count = 0
    for _ in range(20):
        roll = random.randint(1, 6)
        if roll % 2 == 0:  # Check if the roll is an even number
            even_count += 1
    return even_count > 5

result = one_trial()
print(result)

if result > 5:
    print("True")
else:
    print("False")
```



Repeat the function `one_trial` for $N = 100,000$ times and sum up the outcomes and store the result in a variable named `desired_output`. Compute the probability of getting more than 5 times of even numbers by using relative frequency.


```python
import random

def one_trial():
    even_count = 0
    for _ in range(20):
        roll = random.randint(1, 6)
        if roll % 2 == 0:
            even_count += 1
    return even_count > 5

def desired_output():
    trial_successes = 0
    for _ in range(100000):
        if one_trial():
            trial_successes += 1
    return trial_successes

monte_carlo_successes = desired_output()
print(f"Number of trials with more than 5 even numbers: {monte_carlo_successes}")

# Compute the probability of getting more than 5 times of even numbers
N = 100000
probability = monte_carlo_successes / N
print(f"Probability of getting more than 5 even numbers (Monte Carlo): {probability}")
```
